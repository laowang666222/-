# Movies Dataset数据集探索和预处理

## 1 导入依赖库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2 导入数据集

```python
df = pd.read_csv('movies_dataset.csv')
```

对该数据集的id属性进行分析，查看是否有重复

```python
id_count = df["id"].value_counts()
id_count
```

可看出该数据集有大量重复数据，原因为电影浏览次数等相关数据不断更新，同一电影数据多次添加到数据集中，因此对同一电影重复数据进行删除，保留最后一次更新的数据

```python
df=df.drop_duplicates(subset='id', keep='last', inplace=False)
df.info()
```

## 4. 数据缺失的处理

```python
df.info()
```

```python
print(df.isnull().sum())
```

该数据集的数据缺失可能是由于在往数据集中添加数据时发生了遗漏

由于该数据集的特性，采取将将缺失部分剔除的策略，以IMDb-rating属性为例：

```python
df=df.dropna(subset=['IMDb-rating'])
df.info()
```

```python
print(df.isnull().sum())
```

处理后，其余各有主要缺失的属性的缺失值均有一定减少，且appropriate_for，director，run_time，writer的缺失部分与IMDb-rating属性的缺失部分的重合部分超过400个。若对缺失最多的appropriate_for属性的缺失部分剔除：

```python
df=df.dropna(subset=['appropriate_for'])
df.info()
```

```python
print(df.isnull().sum())
```

则IMDb-rating不再有额外缺失，而director，language，run_time，storyline，writer属性依然有额外缺失。
